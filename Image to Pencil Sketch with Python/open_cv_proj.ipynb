**Image to Pencile Sketch with Python**

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('antonio.jpg')
cv2_imshow(img)
```

```python
color_con=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2_imshow(color_con)
```

```python
gray_scale=cv2.cvtColor(color_con,cv2.COLOR_RGB2GRAY)
cv2_imshow(gray_scale)
```

```python
#inverting the image
invert=cv2.bitwise_not(gray_scale)
cv2_imshow(invert)
```

```python
#Gaussian Blur
blur=cv2.GaussianBlur(invert,(15,15),sigmaX=0,sigmaY=0)
cv2_imshow(blur)
```

```python
#pencile_sketch
pencile_sketch=cv2.divide(gray_scale,255-blur,scale=256)
cv2_imshow(pencile_sketch)
```
